```python
lamda_values=pca_out.explained_variance_ratio_
lamda_values
```

```python
#Select the number of main components
def select(pow=0.85):
    sum_of_lamdas=0
    for i in range(len(df.columns)-1):
        sum_of_lamdas=sum_of_lamdas+lamda_values[i]
        if sum_of_lamdas>=pow:
            return(i+1)
print(select())
```

```python
#Remove extra components from dataframe
de_label=[]
for i in range((len(df.columns)-1)-select()):
    current_pc = "PC" + str((len(df.columns)-1)-i)
    de_label.append(current_pc)

df_new_f = df_new.drop(labels=de_label, axis=1)
df_new_f
```
